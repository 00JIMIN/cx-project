```python
df = (df
      .replace({'성별': {'남자': 'M', '여자': 'F'}})  # 성별 변환
      .assign(대학교=lambda a: a['대학교'].str.replace('학교', '', regex=False))  # 대학교에서 '학교' 제거
      .assign(학년=lambda b: b['학년'].apply(
          lambda c: int(c.replace('학년', '').strip()) 
                    if '학년' in c 
                    else ('졸업' if c in ['졸업유예 / 졸업'] 
                          else ('대학원' if c == '대학원생' 
                                else c))))  # 학년에서 '학년' 제거 및 정수형으로 변환
      .replace({'공강유무': {'있다.': 'O', '없다.': 'X'}})  # 공강유무 변환
      .assign(어디가는지=lambda d: d['어디가는지'].str.replace('(교내 내 개방된 공간)', '', regex=False).str.strip())  # '(교내 내 개방된 공간)' 문자열 제거
      .loc[~df['대학교'].str.contains('[a-zA-Z]', na=False)]  # 영어가 포함된 대학교 행 제거 + 노가다
     )

print(df.head())
```

# 현재의 대학생들의 공강시간(설문조사)

#### 뭘 하고 싶은지 묻는 항목 제외함

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import re
import numpy as np

# 폰트 설정 (윈도우의 경우 Malgun Gothic)
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호 깨짐 방지

# 필요한 컬럼만 추출
columns_to_use = ['뭐하는지', '어디가는지', '누구와가는지']
text_data = df[columns_to_use].astype(str)

# '0' 값을 무시하고 결합
text_data = text_data.replace('0', '')  # '0' 값을 빈 문자열로 대체
processed_text = ' '.join(text_data.apply(lambda x: ' '.join(x), axis=1))

# 단어 단위로 분리하기 위해 정규 표현식을 사용
words = re.findall(r'\b\w+\b', processed_text)  # \b는 단어 경계를 나타냄

# 단어 빈도수 계산
word_freq = Counter(words)
```

```python
# 랜덤 시드 설정
np.random.seed(41)  

# 워드클라우드 생성
wordcloud = WordCloud(
    font_path=font_path,  # 한글 폰트 경로 지정
    width=800,
    height=400,
    background_color='white',
    random_state=41  # 랜덤 상태를 설정합니다.
).generate_from_frequencies(word_freq)

# 워드클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 상위 10개 단어를 막대그래프로 시각화
top_words = pd.DataFrame(word_freq.items(), columns=['word', 'frequency']).nlargest(10, 'frequency')

plt.figure(figsize=(10, 5))
plt.bar(top_words['word'], top_words['frequency'], color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# 파이그래프 생성: 빈도수에 따라 내림차순 정렬
top_words_sorted = top_words.sort_values(by='frequency', ascending=False)

# 가장 큰 값의 인덱스를 찾아서 startangle 설정
max_index = top_words_sorted['frequency'].idxmax()
num_words = len(top_words_sorted)
start_angle = (360 / num_words * top_words_sorted.index.get_loc(max_index)) -90

plt.figure(figsize=(8, 8))
plt.pie(top_words_sorted['frequency'], labels=top_words_sorted['word'], autopct='%1.1f%%', startangle=start_angle)
plt.show()
```

```python
# 상위 10개 단어의 빈도 및 비율 계산
total_count = sum(word_freq.values())
top_10_count = top_words['frequency'].sum()

print("상위 10개 단어의 빈도수:")
for index, row in top_words.iterrows():
    word = row['word']
    freq = row['frequency']
    print(f" {word} : {freq}")

print(f"\n전체 단어 수: {total_count}")
```

```python

filtered_df = df[df['충분한지'] == '충분하지 않다.']

university_counts = filtered_df['대학교'].value_counts()

university_counts_df = university_counts.reset_index()
university_counts_df.columns = ['대학교', '빈도수']

print(university_counts_df.head(10))  # 그냥 데이터 많은 곳만 보여주는건가, 학교 이름이 통일이 안되어서 그런가.
```

```python
filtered_df = df[['성별', '누구와가는지']]

filtered_df = filtered_df[filtered_df['누구와가는지'] != '0']  # '0' 값을 제외
gender_who_counts = filtered_df.groupby('성별')['누구와가는지'].value_counts().unstack().fillna(0)

for gender, counts in gender_who_counts.iterrows():
    # 성별 포맷팅
    gender_label = '여성' if gender == 'F' else '남성'
    print(f"{gender_label}의 공강시간은?:")
    sorted_counts = counts[counts > 0].sort_values(ascending=False)  # 0 값 제외
    if sorted_counts.empty:
        print("  데이터 없음")
    else:
        for item, count in sorted_counts.items():
            print(f"  {item}: {int(count)}명")
    print("\n")
```
